```python
import numpy as np

classes=['Python','Java','C++','ML','AI']
class1_student=[3,25,20,15,10]
class2_student=[40,5,10,20,15]
class_index=np.arange(len(classes))
width=0.1
plt.bar(classes,class1_student,width,label='Class1')
plt.bar(class_index+width,class2_student,width,label='Class2')
#xticks means exchange x-axis labels
plt.xticks(class_index+width,classes)
plt.ylabel('Number of students')
plt.xlabel('Classes')
plt.title('Number of students in each class')
plt.title('')

```

```python

x_axis=np.array([0,5,15,15,20,25])
plt.plot(x_axis,x_axis**2,'red',label='x=y**2')
plt.legend()
plt.show()
```

```python
y_axis = x_axis**2
empty_canvas = plt.figure()

create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b')

#empty_canvas = plt.figure()

create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")

create_axis.legend(loc=0)
```
