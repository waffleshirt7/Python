```python
df = pd.read_csv('data.csv', on_bad_lines='skip')
print(df.head())
#head is used to print the rows
#python considers strings as objects

print("\n",df.head(2))
print("\n",df.shape)
print("\n",df.dtypes)
print("\n",df.describe())
```

```python
#Transformation
df['address']=df['address'].str.upper()
print(df)

df.columns=df.columns.str.strip() #Remove leading and trailing spaces
df['age']=(df['age']-df['age'].mean())/df['age'].std()
print(df)
```

```python
df.isnull().sum()

#Checking for duplicates

print(df.duplicated().sum())
#Removing duplicates
df=df.drop_duplicates()
```

```python
#Rename

df=df.rename(columns={'Name':'how old'})
print(df)
```

```python
#Sorting/Describing Data
df=df.sort_values('Attack',ascending=False)
print(df)
```

```python
df=df.sort_values(['Type 1','HP'],ascending=[1,0])
print(df)
```
